# Data Preprocessing of IPL Match dataset:


```python
df[df['city'].isnull()]
```

```python
df['city'].unique()
```

### Replacing Bengaluru with Bangalore:

```python
df['city'].replace(to_replace='Bengaluru',value='Bangalore',inplace=True)
```

```python
for idx in df[df['city'].isna()].index:
        df.loc[idx, 'city'] = 'Sharjah' if df.loc[idx, 'venue'] == 'Sharjah Cricket Stadium' else 'Dubai'
```

```python
df['city'].isnull().sum()
```

```python
df[df['player_of_match'].isnull()]
```

### removing player_of_match, winner, result , result_margin

```python
df.drop([241,486,511,744],inplace=True)
```

```python
df.isnull().sum()
```

```python
df[df['result_margin'].isnull()]
```

```python
df['result_margin'].fillna(0,inplace=True)
```

```python
df[df['method'].isnull()]
```

```python
df['method'].unique()
```

```python
df['method'].fillna('normal',inplace=True)
```

```python
df.isnull().sum()
```

#### Replacing the repeated team :

```python
df['winner'].replace(to_replace='Rising Pune Supergiant',value='Rising Pune Supergiants',inplace=True)
```

# All the null values of  IPL Match dataset are fulfilled
